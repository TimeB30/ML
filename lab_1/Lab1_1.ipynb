# 1. Numpy

1. Создайте матрицу $3 × 3 $ со значениями от 0 до 8.

2. Создайте матрицу шахматной доски  $8 × 8 $, заполненную 'B' и 'W', используя функцию tile.

5. Создайте случайный вектор размером 15 и замените максимальное значение на $NaN$.


```python
vector = np.random.rand(15)
print(vector)
max_index = np.argmax(vector)
vector[max_index] = np.nan
print(vector)
```

6. Рассмотрим массив размерности $(6, 7, 8)$. Каков индекс $(i,j,k)$ 100-го элемента?

```python
array = np.ones((6, 7, 8))
print(array)
index = np.unravel_index(99, array.shape)
print(index)
```

7. Рассмотрим случайную матрицу  $15 × 2 $, представляющую декартовы координаты, преобразуйте их в полярные координаты.

```python
cartesian_coords = np.random.rand(15, 2) * 10
x = cartesian_coords[:, 0]
y = cartesian_coords[:, 1]
r = np.sqrt(x**2 + y**2)  
theta = np.arctan2(y, x)
polar_coords = np.column_stack((r, theta))
print("Декартовы координаты (x, y):\n", cartesian_coords)
print("\nПолярные координаты (r, θ):\n", polar_coords)
```

8. По двум массивам $x$ и $y$ постройте матрицу Коши $C$
 $(C_{ij} =\frac{1}{x_i - y_j})$

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
C = 1.0 / (x[:, np.newaxis] - y)
print("Матрица Коши:\n", C)
```

9. Рассмотрим матрицу $10 × 3 $, извлеките строки с неравными значениями (например, $[2,2,3]$).

```python
matrix = np.random.randint(1, 5, (10, 3))
unique_counts = np.apply_along_axis(lambda x: len(np.unique(x)), 1, matrix)
unique_rows = matrix[unique_counts >= 2]
print("Исходная матрица:\n", matrix)
print("\nСтроки с неравными значениями:\n", unique_rows)
```

10. Извлеките уникальные строки из двумерного массива.


```python
matrix = np.random.randint(1, 5, (10, 3))
unique_rows = np.unique(matrix, axis=0)
print("Исходная матрица:\n", matrix)
print("\nУникальные строки:\n", unique_rows)
```

# Найдите ответы на вопросы с помощью предложенного датасета:

Файл с датасетом - `olympics_dataset.csv`

```python
import pandas as pd
```

```python
DATA_PATH = "olympics_dataset.csv"
df = pd.read_csv(DATA_PATH)
```

1. Сколько спортсменов участвовали в Летних Олимпийских играх за всё время?

```python
summer_olympics = df[df['Season'] == 'Summer']
unique_athletes = summer_olympics['player_id'].nunique()
print(f"Количество спортсменов, участвовавших в Летних Олимпийских играх: {unique_athletes}")
```

2. Найдите участника с наибольшим количеством медалей

```python
medalists = df[df['Medal'] != 'No medal']
medal_count = medalists.groupby(['player_id', 'Name']).size().reset_index(name='Medal_Count')
top_athlete = medal_count.loc[medal_count['Medal_Count'].idxmax()]
print(f"Спортсмен с наибольшим количеством медалей: {top_athlete['Name']}")
```

3. Какой спорт является самым популярным, то есть в нём участвовало больше всего спортсменов за всё время?

```python
unique_athletes_per_sport = df.drop_duplicates(subset=['player_id', 'Sport'])
sport_counts = unique_athletes_per_sport['Sport'].value_counts()
most_popular_sport = sport_counts.idxmax()
most_popular_sport_count = sport_counts.max()
print(f"Самый популярный спорт: {most_popular_sport}")
print(f"Количество спортсменов: {most_popular_sport_count}")
```

4. Кто был победителем медального зачёта в 2004 году?

```python
medals_2004 = df[(df['Year'] == 2004) & (df['Medal'] != 'No medal')]
medal_counts = medals_2004.pivot_table(index='NOC', columns='Medal', aggfunc='size', fill_value=0)
medal_counts['Total'] = medal_counts.sum(axis=1)
top_country_gold = medal_counts.sort_values(by='Gold', ascending=False).iloc[0]
top_country_total = medal_counts.sort_values(by='Total', ascending=False).iloc[0]
print(f"Страна с наибольшим количеством золотых медалей в 2004 году: {top_country_gold.name}, Золото: {top_country_gold['Gold']}")
print(f"Страна с наибольшим количеством медалей в целом в 2004 году: {top_country_total.name}, Всего медалей: {top_country_total['Total']}")
```

5. В каком виде спорта США завоевало наибольшее число медалей?

```python
usa_medals = df[(df['NOC'] == 'USA') & (df['Medal'] != 'No medal')]
sport_medal_count = usa_medals['Sport'].value_counts()
top_sport = sport_medal_count.idxmax()
top_sport_count = sport_medal_count.max()
print(f"Вид спорта, в котором США завоевало наибольшее число медалей: {top_sport}")
print(f"Количество медалей: {top_sport_count}")
```

# С помощью графика покажите:

6. Сколько женщин и мужчин участвовало в Летних Олимпийских играх каждый год?

```python
import matplotlib.pyplot as plt
summer_olympics = df[df['Season'] == 'Summer']
unique_athletes = summer_olympics.drop_duplicates(subset=['player_id', 'Year', 'Sex'])
athlete_counts = unique_athletes.groupby(['Year', 'Sex']).size().unstack()
athlete_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Количество мужчин и женщин, участвовавших в Летних Олимпийских играх по годам')
plt.xlabel('Год')
plt.ylabel('Количество участников')
plt.legend(title='Пол', labels=['Мужчины', 'Женщины'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

7. Какие 10 стран на текущий момент имеют наибольшее количество золотых медалей?


```python
gold_medals = df[df['Medal'] == 'Gold']
gold_medal_count = gold_medals['NOC'].value_counts()
top_10_countries = gold_medal_count.head(10)
plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='bar', color='gold')
plt.title('Топ-10 стран с наибольшим количеством золотых медалей')
plt.xlabel('Страна (NOC)')
plt.ylabel('Количество золотых медалей')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

8. Сколько золотых, серебряных и бронзовых медалей зарабатывал СССР отдельно по каждым играм?

```python
ussr_medals = df[df['NOC'] == 'URS']
medal_counts = ussr_medals.groupby(['Year', 'Medal']).size().unstack(fill_value=0)
medal_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['gold', 'silver', 'brown'])
plt.title('Количество медалей СССР на Олимпийских играх по годам')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.xticks(rotation=45)
plt.legend(title='Тип медали', labels=['Золотые', 'Серебряные', 'Бронзовые'])
plt.tight_layout()
plt.show()
```

9. Сколько медалей завоёвывали спортсмены и спортсменки (отдельно) команды Китая на каждой олимпиаде?


```python
china_medals = df[df['NOC'] == 'CHN']
unique_china_medals = china_medals.drop_duplicates(subset=['player_id', 'Year', 'Sex'])
medal_counts = unique_china_medals.groupby(['Year', 'Sex']).size().unstack(fill_value=0)
medal_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['lightblue', 'salmon'])
plt.title('Количество медалей спортсменов и спортсменок команды Китая на Олимпийских играх')
plt.xlabel('Год')
plt.ylabel('Количество медалей')
plt.xticks(rotation=45)
plt.legend(title='Пол', labels=['Мужчины', 'Женщины'])
plt.tight_layout()
plt.show()
```

10. Какие 10 стран имеют наибольшее количество участников, не получивших никаких медалей ни на одной из Летних олимпиад?

```python
no_medals = df[(df['Season'] == 'Summer') & (df['Medal'] == 'No medal')]
unique_no_medals = no_medals.drop_duplicates(subset=['player_id', 'NOC'])
no_medal_counts = unique_no_medals['NOC'].value_counts()
plt.figure(figsize=(10, 6))
top_10_no_medals = no_medal_counts.head(10)
top_10_no_medals.plot(kind='bar', color='lightcoral')
plt.title('Топ-10 стран с наибольшим количеством участников без медалей на Летних Олимпийских играх')
plt.xlabel('Страна (NOC)')
plt.ylabel('Количество участников без медалей')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
